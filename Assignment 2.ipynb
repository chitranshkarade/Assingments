```python
# 10. How can you tell the difference between break and continue?
# Ans .The break statement in Python terminates the current loop and resumes execution at the next statement
for i in range(5):
    if(i==3):
        break
    print(i)
    
print("this is how Break statement works")
# The continue statement is used to end the current iteration in a for/while loop , and continues to the next iteration.
for i in range(5):
    if(i==3):
        continue
    print(i)
print("this how continue works")    
```
