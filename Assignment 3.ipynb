```python
13. What can you do to save a programme from crashing if it encounters an error?
Ans. To save a programme from crashing if it encounters an error. we can use Exceptional handling/logging.
```

```python
def add(a, b):
    try:
        return a + b
    except Exception as e:
        return 'its an error -  ' + str(e)
 
print(add('', 10))
```
